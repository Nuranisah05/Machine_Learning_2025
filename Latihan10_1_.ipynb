

*   Melakukan Ecoding Dataset


```python
encoder = LabelEncoder()
df['Persepsi Marry'] = encoder.fit_transform(df['Persepsi Marry'])
```

```python
# df = df.replace(LabelEncoder)
df.head()
```

```python
# Menentukan Variabel x dan y
X = df[['Temperatur Udara (°C)', 'Kecepatan Angin (km/jam)']]
y = df['Persepsi Marry']
```

```python
# Mencari K terbaik
akurasi = []

for k in range(1, 9):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
    pred = knn.predict(X)
    acc = accuracy_score(y, pred)
    akurasi.append(acc)
    print(f"K = {k}, Akurasi = {acc}")
```

```python
# Menentukan K terbaik
best_k = np.argmax(akurasi) + 1
print("\nNilai K terbaik berdasarkan akurasi:", best_k)
```

```python
# PREDIKSI PERSEPSI MARRY UNTUK (16°C, 3 km/jam)
model = KNeighborsClassifier(n_neighbors=best_k)
model.fit(X, y)

data_test = [[16, 3]]
hasil_encoded = model.predict(data_test)[0]
hasil = encoder.inverse_transform([hasil_encoded])[0]

print("Persepsi Marry pada suhu 16°C dan angin 3 km/jam =", hasil)
```
